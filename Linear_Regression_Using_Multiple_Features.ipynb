```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([[1000,2],[1500,3],[1800,3],[2400,4],[3000,4]])
Y = np.array([180000,250000,270000,320000,360000])
print(X)
print(Y)
```

```python
X_b = np.c_[np.ones((X.shape[0],1)),X]
print(X_b)
```

```python
weights = np.linalg.inv(X_b.T@X_b)@X_b.T@Y
print(weights)
```

```python
alpha = weights[0]
beta_area = weights[1]
beta_room = weights[2]
print("Intercept (alpha)",round(alpha,2))
print("Coefficient for Area",round(beta_area,2))
print("Coefficient for Room", round(beta_room,2))

```

```python
Y_pred = X_b@weights
print(Y_pred)
```

```python
for actual, predicited in zip(Y,Y_pred):
  print("Actual ",round(actual,2),"Predicted",round(predicited,2))
```

```python
new_house = np.array([1,2200,3])
predicted_price = new_house@weights
print("Predicted Price for 2200sq ft and 3 rooms",round(predicted_price))
```

```python
SS_res = np.sum((Y-Y_pred)**2)
SS_total = np.sum((Y-np.mean(Y))**2)
r2 = 1 - (SS_res/SS_total)
print("R2 scores ",round(r2,2))
```
