```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
plt.show()
```

```python
dx = (1,2,3,4,5)
dy = (2,4,6,4,2)
dy2 = (4,2,4,1,5)

plt.plot(dx,dy,"o-")
plt.plot(dx,dy2,"*-")
plt.show()
#plt.savefig("plot.png")
```

```python
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y,label = 'sin(x)',color = 'blue',linestyle='--')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sin(x) Plot")
plt.legend()
plt.grid(False)
plt.show()
```

```python
dx = (1,2,3,4,5)
dy = (2,4,6,4,2)
dy2 = (4,2,4,1,5)

plt.scatter(dx,dy)
plt.scatter(dx,dy2)
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [1200,1350,1600,1800,1750,1650,1850,1900,1350,1450,1600,1900]

plt.figure()
plt.plot(months, sales)
plt.xlabel("Month")
plt.ylabel("Sales ($USD)")
plt.title("Monthly Sales")
plt.show()
```

```python
months = np.arange(1,13)
city1 = np.random.randint(100,200,12)
city2 = np.random.randint(100,200,12)
city3 = np.random.randint(100,200,12)

rainfall_data = np.array([city1,city2,city3])
monthly_avg = np.mean(rainfall_data,axis=0)

plt.plot(months,monthly_avg)
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.title("Monthly Rainfall Data")
plt.show()
```

```python
fruits = ["apple", "banana", "cherry","orange","pineapple"]
quantity = [20,30,10,50,40]

plt.bar(fruits,quantity,color="skyblue")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.title("Fruit Quantity")
plt.show()
```

```python
fig,ax = plt.subplots()
fruits=["apple", "blueberry", "banana", "cherry", "orange"]
counts=[40,100,30, 70, 55]
bar_labels = ['red', 'blue', 'green', 'pink', 'blue']
bar_colors = ['tab:blue', 'tab:orange', 'red', '#4CAF50', 'purple']
ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
ax.set_ylabel("Fruit Supply")
ax.set_title("Fruit Supply by kind and color")
plt.show()
```

```python
products = ["Milk","Coffee","Cake","Cookie","Bread"]
sales = np.random.randint(8,250,size=5)
plt.bar(products,sales)
plt.xlabel("Products")
plt.ylabel("Units of Sold")
plt.title("Coffee Shop")
#plt.savefig()
plt.show()
```
