```python
import matplotlib.pyplot as plt
dx = (1,2,3,4,5,2,4,6,4,2,4,1,5)
a = [0.5,1.5,2.5,3.5,4.5,5.5]
plt.hist(dx,a,rwidth=0.8)
plt.show()
```

```python
#bell shape
import numpy as np
data = np.random.randn(1000)
plt.hist(data,bins=30,color='purple',edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Uniform distribution
#bell shape
import numpy as np
data = np.random.random(1000)
plt.hist(data,bins=30,color='purple',edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
x = np.random.rand(50)
y = 2*x + np.random.normal(0,0.1,50)

plt.scatter(x,y,color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
java = [78,85,92,70,88]
database = [80,82,75,90,95]
web = [72,68,80,77,85]
ai = [65,70,75,60,72]
data = [java,database,web,ai]
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

```python
a = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(a,vert=True,patch_artist=True)
plt.title('Box Plot')
plt.xticks([1,2,3],['std=1','std=2','std=3'])
plt.grid(True)
plt.show()
```

```python
scores = [45,50,52,55,58,60,62,61,68,70,72,74,76,80,85,120]
plt.boxplot(scores)
plt.title('Box Plot')
plt.xlabel('Scores')
plt.ylabel('Marks')
plt.show()
```

```python
labels = ['Python','Java','C++','JavaScript']
counts = [40,25,20,15]
plt.pie(counts,labels=labels,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title('Programming Languages')
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)
fig,axs = plt.subplots(2,2,figsize=(10,6))

axs[0,0].plot(x,np.sin(x))
axs[0,1].plot(x,np.cos(x))
axs[1,0].plot(x,np.tan(x))
axs[1,1].plot(x,np.exp(x))

plt.tight_layout()
plt.show()
```
