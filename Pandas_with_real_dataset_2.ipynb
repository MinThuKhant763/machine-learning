```python
import pandas as pd
import numpy as np
data = {"Name":["Aung","Chaw","Su","JuJu","Hla","Moe","Kay"],
        "City":["Ygn","Kalaw","Ygn","Mdy","Kalaw","Ygn","Mdy"],
        "Age":[25,35,30,28,22,45,32],
        "Salary":[30000,45000,38000,49000,50000,20000,60000]}
df = pd.DataFrame(data)
print(df)
```

```python
df.groupby("City")["Salary"].mean()
```

```python
df.groupby("City")["Name"].count()
```

```python
df.groupby("City")["Age"].max()
```

```python
data1 = {"Name":["Aung","Chaw","Hla","Moe","Kyaw"],
        "Age":[25,np.nan,28,np.nan,35],
        "Salary":[30000,45000,np.nan,50000,np.nan],
        "City":["Ygn","Kalaw",None,"Mdy","Ygn"]}
df1 = pd.DataFrame(data1)
print(df1)
```

```python
#fill data if null data
df1["Age"].fillna(df1["Age"].mean())
```

```python
#drop data if null value
df1.dropna(subset=["Salary"],inplace=True)
print(df1)
```
