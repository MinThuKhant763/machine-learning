```python
import numpy as np
import pandas as pd
s = pd.Series([10,20,30,40,50])
print(s[0])
print(s[3])
print(s[[0,3]])
```

```python
s = pd.Series([100,200,300,400],index=["A","B","C","D"])
print(s["A"])
print(s["C"])
print(s[["A","C"]])
```

```python
print(s.loc["B"])
```

```python
print(s.iloc[1])
```

```python
print(s[1:3])
```

```python
print(s.loc["A":"C"])
```

```python
s = pd.Series([10,20,30,40,50])
filtered = s[s>30]
print(filtered)
```

```python
filtered = s[(s>=20)]
print(filtered)
```

```python
s = pd.Series(["Apple","Banana","Orange","Mango"])
filtered = s[s.isin(["Grape","Cherry"])]
print(filtered)
```

```python
data1 = [1,2,3]
s1 = pd.Series(data1)
data2 = [4,5,6]
s2 = pd.Series(data2)
print(s1+s2)
```

```python
prices = pd.Series([1500,200,500,2000],index=["Milk","Coffee","Bread","Cookies"])
print(prices)
```

```python
new_price = prices*1.10
print(new_price)
```

```python
expensive = prices[prices>=1500]
print(expensive)
```
