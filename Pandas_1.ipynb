```python
import numpy as np
import pandas as pd
data =  [10,20,30,40,50]
s = pd.Series(data)
print(s)
```

```python
s = pd.Series(data,index=["A","B","C","D","E"])
print(s)
```

```python
data = [100,200,300]
s = pd.Series(data,name="Prices")
print(s)
```

```python
#Key --> index
#Value --> series data
data = {"A":10,"B":20,"C":30}
print(data)
s = pd.Series(data)
print(s)
```

```python
marks = {"Kelvin":85,"Helen":92,"Mia":78}
s = pd.Series(marks)
print(s)
```

```python
data = {"A":10,"B":20,"C":30}
s = pd.Series(data,index=["B","C","A"])
print(s)
```

```python
employee = ("John Smith",29,5500.75,True)
print(employee)
s = pd.Series(employee,index=["Name","Age","Salary","Full Time"])
print(s)
```

```python
arr = np.array([10,20,30,40,50])
print(arr)
s = pd.Series(arr)
print(s)
```

```python
arr = np.array(["Apple",1200,1.5,True],dtype=object)
s = pd.Series(arr,index=["Product","Price","Weight","Instock"])
print(s)
```

```python
temperature = np.array([28.5,29.1,30.2,23.0])
s = pd.Series(temperature,index=["Monday","Tuesday","Wednesday","Thursday"])
print(s)
```
