```python
import pandas as pd
df = pd.DataFrame({"Name":["Aung","Su","May","Hla"],
        "Age":[20,21,24,23],
        "Salary":[30000,50000,60000,40000]})
print(df)
```

```python
df[(df["Age"]>=22)&(df["Age"]<=30)]
```

```python
df[df["Salary"]!=30000]
```

```python
df.sort_values("Age")
```

```python
df.sort_values("Salary")
```

```python
df.sort_values("Salary",ascending=False)
```
